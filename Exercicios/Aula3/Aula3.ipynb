## Exercício 01 – Use um gráfico de pizza para retratar os conjuntos de dados a seguir:
(a) Investimentos: Resultados de uma pesquisa que investigou as intenções de investimentos de adultos
americanos em 2013, quando comparadas com o ano de 2012 (Fonte: CNN).

Investir mais em ações – 562 

Não investir – 288

Investir mais em títulos – 144 

Investir o mesmo valor de 2012 – 461

(b) Maratona da cidade de Nova York: Nacionalidade dos vencedores da prova masculina, até 2012
(Fonte: New York Road Runners).


Estados Unidos – 15 
Tanzânia – 1 Grã-Bretanha – 1
Itália – 4 Quênia – 9 Brasil – 2
Etiópia – 2 México – 4 Nova Zelândia – 1
África do Sul – 2 Marrocos – 1


```python
import pandas as pd
import matplotlib.pyplot as plt

dados = [['Investir mais em ações', 562], ['Não investir', 288], [
    'Investir mais em títulos', 144], ['Investir o mesmo valor de 2012', 461]]
tabela = pd.DataFrame(dados, columns=['Investimentos', 'Frequência'])
tabela['Porcentagem'] = (tabela['Frequência'] /
                         tabela['Frequência'].sum()) * 100
display(tabela)
```

```python
# Gráfico de pizza
labels = tabela['Investimentos']
porcentagens = tabela['Porcentagem']
plt.pie(porcentagens, labels=labels, autopct='%.1f%%')
plt.title('Distribuição das intenções de investimentos de adultos')
plt.show()
```

## Exercício 02
 – A representação gráfica do conjunto de dados do exercício 1(b) não é muito amigável,
uma vez que o número de categorias é excessivo. Reorganize o conjunto de dados de forma que as
categorias denotem os continentes em que se localizam os países natais dos vencedores da maratona
de Nova York. A seguir, construa um novo gráfico de pizza. Qual continente exibe o maior número de
vencedores?


```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados originais
data = {
    'Nacionalidade': ['Estados Unidos', 'Tanzânia', 'Grã-Bretanha', 'Itália', 'Quênia', 'Brasil', 'Etiópia', 'México', 'Nova Zelândia', 'África do Sul', 'Marrocos'],
    'Vencedores': [15, 1, 1, 4, 9, 2, 2, 4, 1, 2, 1]
}

# Mapeamento manual de países para continentes
country_to_continent = {
    'Estados Unidos': 'América do Norte',
    'Tanzânia': 'África',
    'Grã-Bretanha': 'Europa',
    'Itália': 'Europa',
    'Quênia': 'África',
    'Brasil': 'América do Sul',
    'Etiópia': 'África',
    'México': 'América do Norte',
    'Nova Zelândia': 'Oceania',
    'África do Sul': 'África',
    'Marrocos': 'África'
}

# Criar o DataFrame
df = pd.DataFrame(data)

# Adicionar uma coluna "Continente" ao DataFrame usando o mapeamento
df['Continente'] = df['Nacionalidade'].map(country_to_continent)

# Agregar o número de vencedores por continente
continent_victories = df.groupby('Continente')['Vencedores'].sum().reset_index()

# Criar o gráfico de pizza
plt.figure(figsize=(8, 8))
plt.pie(continent_victories['Vencedores'], labels=continent_victories['Continente'], autopct='%1.1f%%', startangle=140)
plt.axis('equal')
plt.title('Vencedores da Maratona de Nova York até 2012 por Continente')

# Exibir o gráfico
plt.show()

# Encontrar o continente com o maior número de vencedores
max_continent = continent_victories[continent_victories['Vencedores'] == continent_victories['Vencedores'].max()]
print(f"O continente com o maior número de vencedores é {max_continent['Continente'].values[0]} com {max_continent['Vencedores'].values[0]} vencedores.")
```

Exercício 03 – Use um gráfico de Pareto para retratar os conjuntos de dados a seguir:


(a) Olimpíadas: Número de medalhas obtidas por atletas de cinco países nos Jogos Olímpicos de
Londres, em 2012.

Alemanha
44


Grã-Bretanha
65


Estados Unidos
104


Rússia
82


China
88


```python
import pandas as pd

# Dados
paises = ['Estados Unidos', 'Rússia', 'China', 'Grã-Bretanha', 'Alemanha']
medalhas = [104, 82, 88, 65, 44]

# Criar o DataFrame
df = pd.DataFrame({'País': paises, 'Número de Medalhas': medalhas})

# Ordenar o DataFrame em ordem crescente
df = df.sort_values(by='Número de Medalhas', ascending=True)

# Exibir o DataFrame ordenado
display(df)
```

```python

df['Porcentagem Acumulada'] = (df['Número de Medalhas'].cumsum() / df['Número de Medalhas'].sum()) * 100

# Criar o gráfico de Pareto
fig, ax1 = plt.subplots(figsize=(10, 6))
ax1.bar(df['País'], df['Número de Medalhas'], color='skyblue', alpha=0.7, label='Número de Medalhas')
ax1.set_xlabel('País')
ax1.set_ylabel('Número de Medalhas', color='skyblue')
ax1.tick_params(axis='y', labelcolor='skyblue')

ax2 = ax1.twinx()
ax2.plot(df['País'], df['Porcentagem Acumulada'], color='orange', marker='o', label='Porcentagem Acumulada')
ax2.set_ylabel('Porcentagem Acumulada (%)', color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

plt.title('Gráfico de Pareto: Número de Medalhas nos Jogos Olímpicos de Londres 2012 por País')

# Exibir o gráfico
plt.tight_layout()
plt.show()
```

(b) Medicamentos: Ocorrência de erros na composição de medicamentos, levantados em um estudo com
duração de 2 meses (Fonte: PubMed Central).


Drogas não autorizadas – 27 

Omissão – 54

 Forma incorreta da droga – 2

Período incorreto – 37 

Dose inadequada – 57

 Droga deteriorada – 2

```python
import pandas as pd
import matplotlib.pyplot as plt

dados = {
    "Drogas não autorizadas": 27,
    "Omissão": 54,
    "Forma incorreta da droga": 2,
    "Dose inadequada": 57,
    "Droga deteriorada": 2
}

# Criar o DataFrame
df = pd.DataFrame(list(dados.items()), columns=["Causa", "Quantidade"])

# Ordenar o DataFrame em ordem decrescente pela coluna "Quantidade"
df = df.sort_values(by="Quantidade", ascending=False)

# Calcular a porcentagem acumulada
df["Porcentagem Acumulada"] = (df["Quantidade"].cumsum() / df["Quantidade"].sum()) * 100

# Criar o gráfico de Pareto
fig, ax1 = plt.subplots(figsize=(12, 6))
ax1.bar(df["Causa"], df["Quantidade"], color="skyblue", alpha=0.7, label="Quantidade")
ax1.set_xlabel("Causa")
ax1.set_ylabel("Quantidade", color="skyblue")
ax1.tick_params(axis="y", labelcolor="skyblue")

ax2 = ax1.twinx()
ax2.plot(df["Causa"], df["Porcentagem Acumulada"], color="orange", marker="o", label="Porcentagem Acumulada")
ax2.set_ylabel("Porcentagem Acumulada (%)", color="orange")
ax2.tick_params(axis="y", labelcolor="orange")

plt.title("Gráfico de Pareto: Causas de Problemas com Drogas")
plt.xticks(rotation=45, ha="right")

# Exibir o gráfico
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Dados fictícios (número de alunos por professor e salário médio em milhares de dólares)
alunos_por_professor = [17.1, 17.5, 18.9, 17.1, 20, 18.6, 14.4, 16.5, 13.3, 18.4]
salario_medio_professores = [28.7, 47.5, 31.8, 28.1, 40.3, 33.8, 49.8, 37.5, 42.5, 31.9]

# Criar o DataFrame
df = pd.DataFrame({'Aluno por professor': alunos_por_professor, 'Salário médio professores': salario_medio_professores})

# Ordenar o DataFrame em ordem decrescente pela coluna "Salário médio professores"
df = df.sort_values(by="Salário médio professores", ascending=False)

# Calcular a porcentagem acumulada e armazenar em uma variável
porcentagem_acumulada = (df["Salário médio professores"].cumsum() / df["Salário médio professores"].sum()) * 100

# Criar o gráfico de Pareto
fig, ax1 = plt.subplots(figsize=(10, 6))
ax1.bar(df["Aluno por professor"], df["Salário médio professores"], color="skyblue", alpha=0.7, label="Salário Médio")
ax1.set_xlabel('Número de Alunos por Professor')
ax1.set_ylabel('Salário Médio dos Professores (em milhares de dólares)', color="skyblue")
ax1.tick_params(axis="y", labelcolor="skyblue")

ax2 = ax1.twinx()
ax2.plot(df["Aluno por professor"], porcentagem_acumulada, color="orange", marker="o", label="Porcentagem Acumulada")
ax2.set_ylabel("Porcentagem Acumulada (%)", color="orange")
ax2.tick_params(axis="y", labelcolor="orange")

plt.title('Gráfico de Pareto: Número de Alunos por Professor vs. Salário Médio dos Professores')

# Exibir o gráfico
plt.tight_layout()
plt.legend()
plt.show()
```
