##**Exercício 01 - Vitórias no futebol americano**

Calcule a amplitude, a média, a variância e o desvio padrão para a população dada.

image.png


```
# Construção do conjunto de dados
vitorias = pd.Series([13, 10, 12, 11, 7, 8, 6, 6,
                      10, 7, 12, 4, 6, 5, 2, 2])
```

```
# Cáclulo da amplitude
amp_vit = vitorias.max() - vitorias.min()
print(f'Amplitude = {amp_vit} vitórias.')

# Cálculo da média
media_vit = vitorias.mean()
print(f'Média = {media_vit:.1f} vitórias.')

# Cálculo da variância populacional (ddof=0)
var_vit = vitorias.var(ddof=0)
print(f'Variância = {var_vit:.1f} vitórias^2.')

# Cálculo do desvio padrão populacional (ddof=0)
std_vit = vitorias.std(ddof=0)
print(f'Desvio padrão = {std_vit:.1f} vitórias.')
```

## **Exercício 05 - Carros por domicílio**

Construa uma tabela de distribuição de frequência para os resultados de uma amostra aleatória do número de carros por domicílio em certa região, conforme histograma. A seguir, estime a média e o desvio padrão amostral.

image.png

```
# Construção das colunas iniciais da tabela de distribuição de frequências
carros = pd.DataFrame({'Num_Carros': [0, 1, 2, 3],
                       'Num_Dom': [3, 15, 24, 8]})
display(carros)
```

```
# Coluna valor * frequência
carros['x*f'] = carros['Num_Carros']*carros['Num_Dom']
display(carros)

# Cálculo da média amostral
obs = carros['Num_Dom'].sum()
media_am = carros['x*f'].sum()/obs
print(f'Média = {media_am:.1f} carros.')
```

```
# Coluna valor - média
carros['x-xb'] = carros['Num_Carros'] - media_am

# Coluna (valor - média)^2
carros['(x-xb)^2'] = carros['x-xb']**2

# Coluna (valor - média)^2 * frequência
carros['(x-xb)^2*f'] = carros['(x-xb)^2']*carros['Num_Dom']
display(carros)
```

```
# Cálculo do desvio padrão amostral
desv_pam_am = (carros['(x-xb)^2*f'].sum()/(obs-1))**(1/2)
print(f'Desvio padrão amostral: {desv_pam_am:.1f} carros')
```

## **Exercício 06 - Média de rebatidas**

determine o coeficiente de variação das distribuições de amostras da média de rebatidas de jogadores de beisebol de dois times adversários. Então, compare os resultados.

image.png

```
# Construção do conjunto de dados
times = pd.DataFrame({'Time A': [0.295, 0.310, 0.325, 0.272, 0.256,
                                 0.297, 0.320, 0.384, 0.235, 0.297],
                      'Time B': [0.223, 0.312, 0.256, 0.300, 0.238,
                                 0.299, 0.204, 0.226, 0.292, 0.260]})
display(times)
```

```
# Cálculo das médias, desvios padrão e coeficiente de variação
est_times = pd.DataFrame(times.mean(axis=0), columns=['Média'])
est_times['DesvPad'] = times.std(axis=0, ddof=0)
est_times['CV'] = est_times['DesvPad']/est_times['Média']*100
display(est_times)
```

*Análise:* Os times possuem médias e desvios padrão de rebatidas muito similares, o que resulta em coeficientes de cariação também muito próximos. Isso nos leva a crer que ambos os times tem o mesmo desempenho médio no quesito rebatidas.
